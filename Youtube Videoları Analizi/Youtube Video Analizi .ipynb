```python
baslik_uzunlugu = []
for i in range(0,len(youtube.index)):
    uzunluk = len(youtube["title"][i])
    baslik_uzunlugu.append(uzunluk)

youtube["title_lenght"] = baslik_uzunlugu
```

```python
youtube
```

```python
degerler = []
for i in range(0,len(youtube.index)):
    etiketler = youtube["tags"][i].split("|")
    degerler.append(len(etiketler))
    
youtube["tag_count"] = degerler
    
```

```python
oran = []
for i in range(0,len(youtube.index)):
    sonuc = int([youtube["likes"][i] / youtube["dislikes"][i]])
    oran.append(sonuc)
youtube["oran"] = oran
```

```python
youtube.sort_values("oran", ascending=False)
```
